# Univariate Analysis

```python
a=df.loc[df['species']=='setosa']
b=df.loc[df['species']=='versicolor']
c=df.loc[df['species']=='virginica']
```

```python
df.species.unique()
```

# Univariate Analysis

```python
plt.plot(a['petal_length'],np.zeros_like(a['petal_length']),'-')
plt.plot(b['petal_length'],np.zeros_like(b['petal_length']),'o')
plt.plot(c['petal_length'],np.zeros_like(c['petal_length']),'*')
```

# Bivariate Analysis

```python
sns.FacetGrid(df,hue='species').map(plt.scatter,'sepal_length','petal_width').add_legend()
```

# Multivariate Analysis

```python
sns.pairplot(df,hue='species',size=3,markers=["o","s","d"])
```
